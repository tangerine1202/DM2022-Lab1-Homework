```python
def read_data(path, category):
  df = pd.read_csv( path, delimiter='\t', header=None, names=['text', 'sentiment'])
  df['text'] = df['text'].str.strip()
  df['category'] = category
  return df
```

```python
amazon_df = read_data('./data/sentiment labelled sentences/amazon_cells_labelled.txt', 'amazon')
imdb_df = read_data('./data/sentiment labelled sentences/imdb_labelled.txt', 'imdb')
yelp_df = read_data('./data/sentiment labelled sentences/yelp_labelled.txt', 'yelp')

# Combine the dataframes
df = pd.concat([amazon_df, imdb_df, yelp_df], ignore_index=True)

print(df.info())
df.head(3)
```

### 5.3 Feature subset selection

```python
count_vectorizer = CountVectorizer()
df_cnt_vect = count_vectorizer.fit_transform(df['text'])
df_cnt_vect = pd.DataFrame(df_cnt_vect.toarray(), columns=count_vectorizer.get_feature_names_out(), index=df.index)

print(df_cnt_vect.shape)
df_cnt_vect.head(3)
```

```python
tmp = df_cnt_vect * np.log((df_cnt_vect.shape[0] + 1) / (df_cnt_vect.astype(bool).sum(axis=0) + 1))
cidx = tmp.sum(axis=0).sort_values(ascending=False).index
ridx = tmp.sum(axis=1).sort_values(ascending=False).index

plt.figure(figsize=(8, 8))
sns.heatmap(
  tmp.loc[ridx, cidx].iloc[:20, :20],
  cmap='Blues', 
  annot=True)

plt.yticks(rotation=0)
plt.tight_layout()
```

### TF-IDF & term-doc

```python
count_vectorizer = CountVectorizer()
tfidf_vectorizer = TfidfVectorizer()

X_count = count_vectorizer.fit_transform(df['text'])
X_tfidf = tfidf_vectorizer.fit_transform(df['text'])
X_count = pd.DataFrame(X_count.toarray(), columns=count_vectorizer.get_feature_names_out(), index=df.index)
X_tfidf = pd.DataFrame(X_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out(), index=df.index)
y_count = df['sentiment']
y_tfidf = df['sentiment']

df_count = X_count.copy()
df_tfidf = X_tfidf.copy()
df_count['label_sentiment'] = y_count
df_tfidf['label_sentiment'] = y_tfidf
df_count['category'] = df['category']
df_tfidf['category'] = df['category']

print(X_count.shape)
print(X_tfidf.shape)
print(y_count.shape)
print(y_tfidf.shape)
```

### Naive Bayes Classifier

```python
def evaluate(X, y, cv=5, random_state=0):
  estimator = BernoulliNB()
  scores = cross_val_score(estimator, X, y, cv=5, scoring='accuracy')
  return scores
```

```python
score_count = evaluate(X_count, y_count)
score_tfidf = evaluate(X_tfidf, y_tfidf)
print('count score (mean/std):', score_count.mean(), score_count.std())
print('tfidf score (mean/std):', score_tfidf.mean(), score_tfidf.std())
```
